```python
#Train & Test Split
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X1,Y, test_size = 0.2, random_state = 123)
print(X_train.shape) 
print(y_train.shape) 
print(X_test.shape)  
print(y_test.shape)
```

```python

from sklearn import linear_model
from sklearn.tree import DecisionTreeRegressor 

regr = DecisionTreeRegressor()
regr.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print(r2_score(y_train, y_train)) #Coefficient of Determination
print(mean_squared_error(y_train, y_train))
```

```python
from sklearn.model_selection import cross_val_score 
accuracies = cross_val_score(estimator = regr, X = X1,\
     y = Y, cv = 15)
print("Accuracy Mean {} Accuracy Variance \
     {}".format(accuracies.mean(),accuracies.std()))
```
